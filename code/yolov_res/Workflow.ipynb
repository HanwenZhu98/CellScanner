```python
import detect
import detect_yocls
import argparse

def Yores_detect(
    source,
    yolo_weight,
    resnet_weight = None,
    name = 'exp',
    save_txt= True,
):
    #Âè™Áî®yolo
    if resnet_weight is None:
        ROOT = detect.ROOT
        detect.run(
        weights=yolo_weight,  # model.pt path(s)
        source=source,  # file/dir/URL/glob, 0 for webcam
        data=ROOT / 'data/coco128.yaml',  # dataset.yaml path
        imgsz=(640, 640),  # inference size (height, width)
        conf_thres=0.4,  # confidence threshold
        iou_thres=0.4,  # NMS IOU threshold
        max_det=1000,  # maximum detections per image
        device='',  # cuda device, i.e. 0 or 0,1,2,3 or cpu
        view_img=False,  # show results
        save_txt=save_txt,  # save results to *.txt
        save_conf=True,  # save confidences in --save-txt labels
        save_crop=False,  # save cropped prediction boxes
        nosave=False,  # do not save images/videos
        classes=None,  # filter by class: --class 0, or --class 0 2 3
        agnostic_nms=False,  # class-agnostic NMS
        augment=False,  # augmented inference
        visualize=False,  # visualize features
        update=False,  # update all models
        project=ROOT / 'runs/detect1',  # save results to project/name
        name = name,  # save results to project/name
        exist_ok=False,  # existing project/name ok, do not increment
        line_thickness=1,  # bounding box thickness (pixels)
        hide_labels=True,  # hide labels
        hide_conf=False,  # hide confidences
        half=False,  # use FP16 half-precision inference
        dnn=False,  # use OpenCV DNN for ONNX inference
        )     
    #both model used
    else:
        ROOT = detect_yocls.ROOT
        save_dir = detect_yocls.run(
        weight_path = None,
        weights_yolo = yolo_weight,
        source=source, 
        weights_cls= resnet_weight,
        data=ROOT / 'data/coco128.yaml',  # dataset.yaml path
        imgsz=(640, 640),  # inference size (height, width)
        conf_thres=0.4,  # confidence threshold
        iou_thres=0.4,  # NMS IOU threshold
        max_det=1000,  # maximum detections per image
        device='',  # cuda device, i.e. 0 or 0,1,2,3 or cpu
        view_img=False,  # show results
        save_txt=save_txt,  # save results to *.txt
        save_conf=True,  # save confidences in --save-txt labels
        save_crop=False,  # save cropped prediction boxes
        nosave=False, # do not save images/videos
        classes=None,  # filter by class: --class 0, or --class 0 2 3
        agnostic_nms=False,  # class-agnostic NMS
        augment=False,  # augmented inference
        visualize=False,  # visualize features
        update=False,  # update all models
        project=ROOT / 'runs/detect1',  # save results to project/name
        name=name,  # save results to project/name
        exist_ok=False,  # existing project/name ok, do not increment
        line_thickness=1,  # bounding box thickness (pixels)
        hide_labels=True,  # hide labels
        hide_conf=False,  # hide confidences
        half=False,  # use FP16 half-precision inference
        dnn=False,  # use OpenCV DNN for ONNX inference
        save_summary = False
        )
```

```python
img_source = '/cluster/home/hanwen/a_Celldetect/idp_dataset/images/test/'
model_dir = '/cluster/home/hanwen/a_Celldetect/models/Models/'
resnet_weight = '/cluster/home/hanwen/a_Celldetect/models/Models/res_weight_2022216/best.pt'


import os
for model_id in os.listdir(model_dir):
    print(model_id)
    if model_id.startswith('res') or model_id.startswith('.'):
        continue
    else:
        if model_id.endswith('2'):
            Yores_detect(
                source = img_source,
                yolo_weight=model_dir + '{}/weights/best.pt'.format(model_id),
                name = 'idp1_detect_'+model_id
            #     resnet_weight='/Users/zhuhanwen/Desktop/project/Celldetectproject/project_summary/Models/weight_2022216/best.pt',
            )
        else:
            Yores_detect(
                source = img_source,
                yolo_weight=model_dir + '{}/weights/best.pt'.format(model_id),
                resnet_weight= resnet_weight,
                name = 'idp1_detect_'+model_id+'_res'
            #     resnet_weight='/Users/zhuhanwen/Desktop/project/Celldetectproject/project_summary/Models/weight_2022216/best.pt',
            )
```

```python
#pred& 2 val
import val_map
ROOT = val_map.ROOT
import os

class opt_:
    def __init__(self, 
                 data= './data.yaml',
                 weights=ROOT / 'yolov5s.pt', 
                 batch_size=1, 
                 imgsz=640, 
                 conf_thres=0.4, 
                 iou_thres=0.4, 
                 task='val', 
                 device='', 
                 workers=0,
                 single_cls=False, 
                 augment=False,  
                 verbose=False,
                 save_conf=True, 
                 save_hybrid=False, 
                 save_json=False, 
                 save_txt=False, 
                 project=ROOT / 'runs/val_0326_1',
                 name='exp', 
                 exist_ok=False, 
                 half=False, 
                 dnn=False,
                 label_path = None,
                 shift = False,
                ):
        self.augment = augment
        self.batch_size = batch_size
        self.conf_thres = conf_thres
        self.data = data
        self.device = device
        self.dnn = dnn
        self.exist_ok = exist_ok
        self.half = half
        self.imgsz = imgsz
        self.iou_thres = iou_thres
        self.name = name
        self.project = project
        self.save_conf = save_conf
        self.save_hybrid = save_hybrid
        self.save_json = save_json
        self.save_txt = save_txt
        self.single_cls = single_cls
        self.task = task
        self.verbose = verbose
        self.weights = weights
        self.workers = workers
        self.label_path = label_path
        self.shift = shift
```

```python
res_in_detect = '/cluster/home/hanwen/a_Celldetect/runs/detect0326_idp/'
res_summary = {}
for m in [_ for _ in os.listdir(res_in_detect) if not _.startswith('.')]:
    lab_path = res_in_detect+m+'/labels/'

    opt = opt_(weights='/cluster/home/hanwen/a_Celldetect/models/Models/tm2/weights/best.pt',
              label_path = lab_path,
               name = m,
                shift = True)
    res_summary[m] = val_map.main(opt)
```
